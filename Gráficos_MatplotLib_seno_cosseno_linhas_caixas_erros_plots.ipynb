```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
a = np.linspace(0, 4*np.pi, 1000)
b = np.sin(a)
```

```python
a
```

```python
b
```

```python
fig, ax= plt.subplots()
ax.plot(a,b)
fig.show()
```

```python
fig, ax= plt.subplots()
a = np.random.uniform(0, 1, 1000)
b = np.random.uniform(0, 1, 1000)
ax.scatter(a, b)

```

```python
fig, ax= plt.subplots()
a = np.arange(15)
b = np.random.uniform(1, 15, 15)
ax.bar(a, b)
```

```python
fig, ax= plt.subplots()
c = np.random.uniform(0, 1, (7,7))

ax.imshow(c)
```

```python
fig, ax= plt.subplots()
d = np.random.uniform(0, 1, (9,9))

ax.contourf(d)
```

```python
fig, ax= plt.subplots()
e = np.random.uniform(0, 1, 4)

ax.pie(e);
```

```python
fig, ax= plt.subplots()
z = np.arange(5)
y = np.random.uniform(0, 1, 5)
ax.errorbar( z, y, y/4)
```

```python
fig, ax= plt.subplots()

z = np.random.normal(0, 1, (100,5))

ax.boxplot(z);
```

```python
fig, ax= plt.subplots()
x = np.linspace(0, 10, 100)
y = np.sin(x)
ax.plot(x, y, color='red');
```

```python
fig, ax= plt.subplots()
x = np.linspace(0, 10, 100)
y = np.cos(x)
ax.plot(x, y, color='green');
```
